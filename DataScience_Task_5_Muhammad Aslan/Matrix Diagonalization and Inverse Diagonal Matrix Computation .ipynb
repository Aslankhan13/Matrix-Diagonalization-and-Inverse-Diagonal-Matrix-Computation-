```python
import numpy as np

# Step 1: Define the new matrix A
A = np.array([[6, 2],
              [2, 3]])

# Step 2: Compute the eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)

# Step 3: Construct the diagonal matrix D from the eigenvalues
D = np.diag(eigenvalues)

# Step 4: The eigenvector matrix P is already obtained from eigenvectors
P = eigenvectors

# Step 5: Compute the inverse of the diagonal matrix D
D_inv = np.linalg.inv(D)

# Output the results
print("Matrix A:")
print(A)

print("\nEigenvalues (Diagonal elements of D):")
print(eigenvalues)

print("\nEigenvectors (Columns of P):")
print(eigenvectors)

print("\nDiagonal Matrix D:")
print(D)

print("\nInverse of Diagonal Matrix D_inv:")
print(D_inv)
```
